```python
def get_seq_sim(mouse_p, human_p):
    
    f = open("data/mouse-human.blast")
    blast = f.readlines()
    f.close()
    
    count=0
    pairs = []
    scores = []
    
    for i in range(len(blast)):
        tokens = blast[i].strip("\n").split("\t")
        if tokens[0] in mouse_p and tokens[1] in human_p and float(tokens[2]) > 100: # change value for diff alignments
            pairs.append(tokens[0]+","+tokens[1])
            scores.append(float(tokens[2]))
            
    return pairs, scores
        
```
